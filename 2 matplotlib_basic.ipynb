```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50]
```

```python
plt.plot(x)
plt.show()
```

```python
y = [100,200,300,400,500]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,c = 'red')
plt.show()
```

```python
plt.scatter(x,y, c = ['red','black','blue','green','yellow'])
plt.xlabel('Value of X')
plt.ylabel('Valur of Y')
plt.title('My first scatter plot')

plt.show()
```

```python
import numpy as np
```

```python
np.pi
```

```python
x = np.linspace(np.pi, -np.pi, 200, endpoint=True)
```

```python
len(x)
```

```python
s = np.sin(x)
ss = np.cos(x)
color = np.random.rand(200)
size = 50*np.random.rand(200)
```

```python
plt.plot(x,s, label = 'Sin', c = 'black')
plt.scatter(x,ss, label = 'Cos', c = color, s = size)
plt.grid()
plt.legend(loc = 'upper right')
plt.show
```
